DISTRIBUTION 

```python
# Normal distribution 

import matplotlib.pyplot as plt 
# define normal distribution data
normal_data = np.random.normal(loc=0, scale=1, size=1000)


plt.subplot(3, 1, 1)
plt.hist(normal_data, bins = 30, color = "lightblue",alpha = 0.7)
plt.title("Normal disribution")
plt.xlabel("value")
plt.ylabel("frequency")
plt.tight_layout()
plt.show()

# uniform distribution
# define uniform distribution data
Uniform_data = np.random.uniform(low=0, high=10, size=1000)
plt.subplot(3, 1, 2)
plt.hist(Uniform_data, bins= 30,color ="lightgreen",alpha = 0.7)
plt.title("Uniform distribution")
plt.xlabel("value")
plt.ylabel("frequency")
plt.tight_layout()
plt.show()

# Exponential distribution plot
# define exponential distribution data
Exponential_data = np.random.exponential(scale=1.0, size=1000)

plt.subplot(3, 1, 3)
plt.hist(Exponential_data, bins= 30,color = "salmon",alpha = 0.7)
plt.title("Exponential  distribution")
plt.xlabel("value")
plt.ylabel("frequency")

plt.tight_layout()
plt.show()
```
